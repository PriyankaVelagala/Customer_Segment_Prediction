## Kmeans

elbow for optimal k

```
texts = df.Description.unique()
#target = dataset.target # Extract target

#from nltk.corpus import stopwords
#stop = list(stopwords.words('english'))
#stop.extend('red pink white'.split())
#vectorizer = TfidfVectorizer(analyzer = 'word',stop_words=set(stop))


from sklearn.feature_extraction import text 
my_additional_stop_words = ['red', 'white', 'pink', 'blue', 'black', 'green']
stop_words = text.ENGLISH_STOP_WORDS.union(my_additional_stop_words)
vectorizer = TfidfVectorizer(analyzer = 'word',stop_words=set(stop_words))


#vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)
```

```
Sum_of_squared_distances = []
K = [1, 50, 100, 200, 300, 400, 500, 600, 800, 1000]
#range(200,201)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)
```

```
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```
Sum_of_squared_distances = []
K = range(1,20)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(X)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```
from yellowbrick.cluster import KElbowVisualizer


model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,22))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```
number_of_clusters = 16

# https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html
model = KMeans(n_clusters=number_of_clusters, 
               init='k-means++', 
               max_iter=100, # Maximum number of iterations of the k-means algorithm for a single run.
               n_init=1)  # Number of time the k-means algorithm will be run with different centroid seeds. The final results will be the best output of n_init consecutive runs in terms of inertia.

model.fit(X)
```

```
Y = model.labels_
Y
```

```
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```

```
for i in range(number_of_clusters):
    print("Cluster %d:" % i),

    for ind in order_centroids[i, :]:
        print(' %s' % terms[ind])
```

# Analysis per cluster

```
df_cluster = df.merge(clustered, how = "left", left_on = "Description", right_on = "Product Description")
df_cluster['total_rev'] = df_cluster['Quantity']*df_cluster['UnitPrice']
df_cluster.head()
```

Top product by revenue

```
t = df_cluster[['cluster', 'Description', 'total_rev']].copy()
t = t.groupby(['cluster','Description']).sum().reset_index()

max_rev_product = pd.DataFrame(columns = ['cluster', 'TopProductName', 'revenue'])


for i in range(0,number_of_clusters):
  temp = t.iloc[t[t.cluster == i].total_rev.idxmax(axis = 0)]
  prod_name = temp.Description
  rev = temp.total_rev
  max_rev_product =  max_rev_product.append({'cluster': i, 'TopProductName': prod_name, 'revenue': rev}, ignore_index = True)
  
max_rev_product
```

Top product by customer popularity

```
s = df_cluster[['cluster', 'Description', 'CustomerID']].groupby(['cluster','Description']).CustomerID.nunique()
s = pd.Series(s).to_frame().reset_index()
s.head()
```

```
pop_product = pd.DataFrame(columns = ['cluster', 'PopProductName', 'numofCust'])


for i in range(0,number_of_clusters):
  temp = s.iloc[s[s.cluster == i].CustomerID.idxmax(axis = 0)]
  prod_name = temp.Description
  numofCust = temp.CustomerID
  pop_product =  pop_product.append({'cluster': i, 'PopProductName': prod_name, 'numofCust': numofCust}, ignore_index = True)
  
pop_product
```

```
prod_rec_by_cluster = max_rev_product.merge(pop_product, how = "inner", left_on = "cluster", right_on = "cluster").drop(columns = ["revenue", "numofCust"])
prod_rec_by_cluster.head()
```

#Customer spending per cluster

```
df_catid = df.merge(clustered, how = "left", left_on = "Description", right_on = "Product Description")
df_catid['total'] = df_catid['UnitPrice']*df_catid['Quantity']
df_catid.head()
```

```
df_catid = df_catid.groupby(['CustomerID', 'cluster']).sum()#['total']#.reset_index()[['CustomerID', 'cluster', 'total']]
df_catid = df_catid.reset_index()
df_catid = df_catid.pivot(index = 'CustomerID', columns = 'cluster', values = 'total').fillna(0)
df_catid
```

```
def top_cluster(x):
  #print(x)
  top_cluster = -1
  top_amt = -1 
  
  for i in range(0,number_of_clusters):
    if x[i] > top_amt:
      top_amt = x[i]
      top_cluster = i 
      
  return top_cluster

df_catid['top_cluster'] = df_catid.apply(top_cluster, axis = 1)
df_catid = df_catid.reset_index()
df_catid.head()
```

```
df_catid = df_catid.merge(prod_rec_by_cluster, how = "left", left_on = "top_cluster", right_on = "cluster").drop(columns = ['cluster'])
df_catid.head()
```
